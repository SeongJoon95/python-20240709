##### 그래프의 종류
**막대그래프** : 데이터를 카테고리 형태별로 값을 막대로 표현 할때  
사용하는 그래프 형식 (수직 막대 그래프, 수평 막대 그래프)
> 카테고리별 데이터를 비교하고자 할때 사용  (제품 판매량, 지역별 인구수)

**선 그래프** : 시간흐름에 따라 데이터 포인터를 선으로  
연결하여 사용하는 그래프 형식
> 하나의 요소에 대해 시간의 흐름에 따라 변경값을 나타내고자 할때 사용(기온, 차트)

**파이 차트** : 전체 데이터에 대한 각 요소의 비율을 원형으로  
나타낼때 사용하는 그래프 형식
> 전체에 대한 요소의 비율을 표현하고자 할때 사용 (투표 결과)

**산점도** : 두 변수 간의 상관관계를 표현할때 사용하는 그래프 형식
> 각 요소간 관계성을 파악할때 사용(키와 몸무게 상관관계 비교)

**히스토그램** : 범위별로 그룹화하여 각 그룹에 속하는 데이터의  
수를 막대로 표현하는 그래프 형식
> 데이터의 분포를 비교할때 사용 (시험 점수 분포)

**영억 그래프** : 선 그래프 아래쪽 빈 공간을 채운 그래프
> 시간의 흐름에 따라 데이터를 비교할 때 사용, 변화량을 강조하고자 할때 사용

**박스 그래프** : 데이터의 분포를 나타낼때 사용하는 그래프 형식
> 최대값, 최소값, 사분위수로 데이터의 이상치를 판단할때 사용

### 선 그래프
가장 기본적인 그래프 형식, 시간의 흐름 및 순서에 따라서  
데이터의 변화량을 보여주고자 할때 사용  
plot 패키지의 plot() 메서드로 간단하게 생성 가능

````python
plt.plot([1, 2, 4, 8, 16])
plt.show()                                                  
````

````python
# 그래프의 제목 지정 title()메서드 사용
# plot(x축에 대한 리스트, y축에 대한 리스트)
plt.title('x 축 추가')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16])
plt.show()
````

````python
# 차트 스타일 지정
# plot() 메서드의 세번째 매개변수로
# 지정된 문자열 포멧으로 전달하여 스탕리 지정가능
plt.title('c2-- 스타일 적용')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 'c2--')
plt.show()
````

````python
# 스타일 매개변수
# color (c) : 선 생상 , linewidth (lw) : 선 굵기 , linestyle (ls) : 선 스타일
# marker : 마커 종류, markersize(ms) : 마커 크기, 
# markeredgecolor (mec) : 마커 테두리 색상, markeredgewidth (mew) : 마커 테두리 굵기
# markerfacecolor (mfc) : 마커 내부 색상
plt.title('c2-- 스타일 적용')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
         c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.show()
````

````python
# 차트 범위 지정
# xlim(), ylim() 메서드로 x축과 y축의 상한 하한을 지정할 수 있음
plt.title('차트 범위 지정')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
         c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xlim(2019,2025)
plt.ylim(3, 20)
plt.show()
````

````python
# 틱 지정
# xticks(), yticks() 메서드로 틱의 위치나 라벨 지정 가능
plt.title('틱 지정')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
         c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점','중간점','고점'])
plt.show()
````

````python
# 그리드
# grid() 메서드로 그리드를 표시할지 결정할 수 있음
plt.title('그리드')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
         c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점','중간점','고점'])
plt.grid(True)
plt.show()
````

````python
# 차트 겹치기
# plot() 메서드를 중복해서 호출하여 차트를 겹칠 수 있음
plt.title('겹쳐 그리기')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
         c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.plot([2020, 2021, 2022, 2023, 2024], [16, 8, 4, 2, 1], 
         c='#88990050', lw=5, ls=':', marker='*', ms=10, mfc='#998800')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점','중간점','고점'])
plt.grid(True)
plt.show()
````

````python
# 범례
# regend() 메서드를 통해 범례를 표시할 수 있음
# loc 매개변수에 위치에 해당하는 정수를 전달하여 위치 지정 가능
plt.title('범례 표시')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
         c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988', label='증감')
plt.plot([2020, 2021, 2022, 2023, 2024], [16, 8, 4, 2, 1], 
         c='#88990050', lw=5, ls=':', marker='*', ms=10, mfc='#998800', label='감소')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점','중간점','고점'])
plt.grid(True)
plt.legend(loc=0)
plt.show()
````

````python
# 각 틱의 라벨
# xlable(), ylabel() 메서드로 x축, y축의 라벨을 표시할 수 있음
plt.title('각 틱 라벨')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
         c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988', label='증감')
plt.plot([2020, 2021, 2022, 2023, 2024], [16, 8, 4, 2, 1], 
         c='#88990050', lw=5, ls=':', marker='*', ms=10, mfc='#998800', label='감소')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점','중간점','고점'])
plt.xlabel('연도')
plt.ylabel('가치')
plt.grid(True)
plt.legend(loc=0)
plt.show()
````

### 막대 그래프
x축의 값이 분류가 가능한 카테고리 형태일때 bar(), barh() 메서드로 막대 그래프를 표현할 수 있음

````python
plt.title('막대 그래프')
plt.bar(['a','b','c'], [20, 78, 43], color='#ff0000')
plt.show()
````

````python
# 가로 막대그래프
# barch() 메서드로 가로 막대그래프를 표현 가능
# 오차 표시
# xerr, yerr 매개변수로 오차를 지정할 수 있음
plt.title('막대 그래프')
plt.barh(['a','b','c'], [20, 78, 43], xerr=[2, 10, 4], color='#ff000050')
plt.show()
````

### 스템 플롯
기준점을 기준으로 오차범위를 표현하고자 할때 사용하는 차트

````python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title('스템 플롯')
plt.stem(x, np.cos(x))
plt.show()
````

### 파이차트
각 분류가 전체에서 차지하는 기준을 표현하고자 할때 사용하는 차트


````python
plt.title('파이 차트')
plt.pie([25, 45, 30], labels=['사과', '바나나', '포도'],
        explode=[0,.1,0], colors=['#ff000050', '#00ff0050','#0000ff50'],
        shadow=True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
````

### 점 그래프
변수간의 상관관계 표현할때 사용하는 그래프

````python
x = np.random.normal(0,1,100)
y1 = np.random.normal(0,1,100)
y2 = np.random.normal(0,1,100)

plt.title('점 그래프')
plt.scatter(x, y1)
plt.grid(True)
plt.show()
````

````python
plt.title('버블 차트')
plt.scatter(x, y1, c=y2)
plt.grid(True)
plt.show()
````

### 컨투어 플롯
3차원 자료에서 x,y 입력과 z출력이 존재할 때, z를 선과 색상으로 표현하는 그래프

````python
def func(x,y):
    z = (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
    return z

x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
xx, yy = np.meshgrid(x, y)
zz = func(xx, yy)

plt.title('등고선 그래프')
plt.contour(xx, yy, zz, colors='black')
plt.contourf(xx, yy, zz)
plt.grid(True)
plt.show()
````

해당 데이터를 그래프로 시각화하려 한다. 

그래프의 제목은 Population (millions) 이고 선의 두깨는 5, 선의 모양은 대쉬, 마커는 점, 마커의 크기는 10으로 지정하며

남자 그래프는 선의 색상은 #0000ff50, 마커의 색상은 #0000ff으로 지정하고

여자 그래프는 선의 색상은 #ff000050, 마커의 색상은 #ff0000으로 지정하하여 아래와 같이 표현하려고 한다.

이때 이 python의 matplotlib의 pyplot을 이용하여 그래프를 그리는 코드를 작성하시오.


````python
plt.title('Population (millions)')
plt.plot([2000, 2005, 2010, 2015, 2020], [23.5, 23.9, 24.3, 24.6, 25.2],
        c='#0000ff50',lw='5', ls=':', marker='.', ms=10,mfc='#0000ff', label='Male')
plt.plot([2000, 2005, 2010, 2015, 2020], [22.7, 23.1, 23.6, 24.0, 24.8],
        c='#ff000050',lw='5', ls=':', marker='.', ms=10,mfc='#ff0000', label='Female')
plt.xticks([2000, 2005, 2010, 2015, 2020])
plt.yticks([23.0, 23.5, 24.0, 24.5, 25.0])
plt.grid(True)
plt.legend(loc=4)
plt.show()
````
